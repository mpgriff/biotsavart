```python
import biotsavart as bs
import matplotlib.pyplot as plt
import numpy as np
```

```python
mu0 = np.pi*4e-7
I = 1.
R = 1.
lp = bs.Loop.circle(R, line=0.1)

z = np.linspace(-5, 5, 1001)
r_obs = bs.ObsPoint(z=z).xyz
```

```python
fig,ax = plt.subplots()
ax.plot(lp.x, lp.y, '.-')
ax.set_aspect('equal')
```

```python
bz = bs.calc_bfield(lp.xyz, r_obs, I=I)[..., 2]
bz_anly = 0.5*mu0*I*R**2 / (R**2 + z**2)**(3/2)

print('they are all close' if np.allclose(bz, bz_anly) else 'not all close')

fig,axs = plt.subplots(2,1)
axs[0].plot(z, bz_anly)
axs[0].plot(z, bz, '--')
axs[1].plot(z, bz_anly-bz)
```
